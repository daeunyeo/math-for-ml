## Concept Summary

- System of linear equations: multiple conditions that must hold simultaneously
- Geometrically: each equation is a line & solution is where they intersect
- Three outcomes: one solution / no solution / infinitely many
- Matrix form: Ax = b (A = coefficients, x = unknowns, b = right-hand side)

## Problems

1. 2x + y = 5, x - y = 1 → add both equations to eliminate y → x = 2, y = 1
2. 3x + 2y = 12, x + y = 5 → subtract second from first → x = 2, y = 3
3. x + y = 3, 2x + 2y = 6 → second is twice the first → infinitely many solutions

```python
# problem 1
import numpy as np

A = np.array([[2,  1],
              [1, -1]])
b = np.array([5, 1])

x = np.linalg.solve(A, b)
print(f"x = {x[0]}, y = {x[1]}") 
```

```python
# problem 2
A2 = np.array([[3, 2],
              [1, 1]])
b2 = np.array([12, 5])

x2 = np.linalg.solve(A2, b2)
print(f"x = {x2[0]}, y = {x2[1]}")  
print(np.allclose(x2, [2, 3]))  # True = matches expected values
```

```python
# problem 3 
A3 = np.array([[1, 1],
               [2, 2]])  
b3 = np.array([3, 6])

try:
    x3 = np.linalg.solve(A3, b3)
except np.linalg.LinAlgError as e:
    print("에러 발생:", e)
```

```python
# 예제 1 시각화
import matplotlib.pyplot as plt

x_vals = np.linspace(-1, 5, 100)
y1 = 5 - 2*x_vals      # 2x + y = 5
y2 = x_vals - 1         # x - y = 1

plt.plot(x_vals, y1, label='2x + y = 5')
plt.plot(x_vals, y2, label='x - y = 1')
plt.scatter(2, 1, color='red', zorder=5, label='(2, 1)')
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.ylim(-3, 7)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

## Reflection Note
- Matrix form Ax = b lets the same operation scale to any number of variables. This is when I understood why linear algebra is the language of ML, not just a math tool.
- Floating point error is expected behavior, not a mistake. Realizing this changed how I read code output entirely.
- np.linalg.solve hides the complexity, but knowing the Ax = b structure underneath makes it more than just a black box.
- A matrix can be read as a collection of column vectors, which reframes what the whole structure represents. The same object, but a completely different way of seeing it. This goes beyond Ch2.1 scope and connects to linear transformations in Ch2.2.
