## Concept Summary

**Groups**
a group G = (G, ⊗) is a set with an operation that satisfies
1. closure: the result of x ⊗ y stays inside G
2. associativity: bracketing order does not change the result
3. neutral element: there exists e such that x ⊗ e = x
4. inverse element: for every x, there exists y such that x ⊗ y = e

if commutativity (x ⊗ y = y ⊗ x) also holds, it is called an abelian group.

**Vector Spaces**

a vector space V = (V, +, ·) has two operations
- inner operation (+): vector + vector → vector
- outer operation (·): scalar × vector → vector

to be a vector space, four conditions must hold
1. (V, +) is an abelian group
2. distributivity: λ(x+y) = λx+λy and (λ+ψ)x = λx+ψx
3. associativity of outer operation: λ(ψx) = (λψ)x
4. neutral element: 1·x = x

elements x ∈ V are called vectors, λ ∈ R are called scalars.

- R^2 (full 2D plane): satisfies all four conditions -> vector space
- x-axis (vectors where y=0): satisfies all four conditions -> vector space
- y=1 line: does not contain the zero vector, so condition 1 fails -> not a vector space

**Vector Subspaces**

a subspace U is a smaller vector space that lives inside V. to confirm U is a subspace, only three conditions need to be checked
1. U contains the zero vector
2. closed under scalar multiplication: scaling any vector in U keeps it in U
3. closed under addition: adding any two vectors in U keeps the result in U

U automatically inherits all other vector space properties from V, so these three are sufficient.

- solution set of Ax=0: zero vector always satisfies it, scaling or adding solutions gives another solution -> subspace
- solution set of Ax=b (b≠0): zero vector does not satisfy Ax=b, and adding two solutions gives A(x+x)=2b which is outside the set -> not a subspace


## Verification

**Vector Space (R^2)**

verified all four conditions using assert: abelian group, distributivity,
associativity of outer operation, neutral element

**Vector Subspace**

Figure 2.6: A fails scalar multiplication closure, B has no zero vector,
C fails scalar multiplication closure with negative scalar, D passes all three

- Ax=0: all three conditions hold -> subspace
- Ax=b: fails zero vector and addition closure -> not a subspace
- intersection of xy-plane and xz-plane (x-axis): all three conditions hold -> subspace

```python
# Vector Space (R^2, +, *)
import numpy as np

x = np.array([1, 2])
y = np.array([3, 4])
lam = 3
psi = 2

# 1. Abelian Group
assert np.array_equal(x + y, np.array([4, 6]))           # closure
assert np.array_equal((x + y) + x, x + (y + x))          # associativity
assert np.array_equal(x + np.array([0, 0]), x)            # neutral element
assert np.array_equal(x + (-x), np.array([0, 0]))         # inverse element
print("1. Abelian group: ok")

# 2. Distributivity
for a, b in [(lam, psi), (psi, lam)]:
    assert np.array_equal(a * (x + y), a * x + a * y)
    assert np.array_equal((a + b) * x, a * x + b * x)
print("2. Distributivity: ok")

# 3. Associativity (outer operation)
assert np.array_equal((a * b) * x, a * (b * x))
print("3. Associativity (outer operation): ok")

# 4. Neutral element (outer operation)
assert np.array_equal(1 * x, x)
print("4. Neutral element (outer operation): ok")
```

```python
# Vector Subspace - Figure 2.6

# A: square { [x,y] | -1 <= x <= 1, -1 <= y <= 1 }
u = np.array([0.8, 0.8])
lam = 3
scaled = lam * u
assert not (np.all(scaled >= -1) and np.all(scaled <= 1))
print("A: not a subspace")

# B: line not through origin
assert not False
print("B: not a subspace")

# C: wedge region with origin as vertex
u = np.array([1, 0.5])
scaled = -1 * u
assert not np.all(scaled >= 0)
print("C: not a subspace")

# D: x-axis { [a, 0] | a ∈ R }
u = np.array([2, 0])
v = np.array([5, 0])
lam = -3
assert np.array_equal(np.array([0, 0])[1], 0)
assert np.array_equal((lam * u)[1], 0)
assert np.array_equal((u + v)[1], 0)
print("D: subspace")
```

```python
# Vector Subspace - Example 2.12

A = np.array([[1, 2, 3],
              [4, 5, 6]])
x1 = np.array([1, -2, 1])
x2 = np.array([2, -4, 2])
lam = 5

# Ax = 0
assert np.array_equal(A @ np.zeros(3), np.zeros(2))
assert np.array_equal(A @ (lam * x1), np.zeros(2))
assert np.array_equal(A @ (x1 + x2), np.zeros(2))
print("Ax=0: subspace")

# Ax = b
b = np.array([1, 2])
x3 = np.linalg.lstsq(A, b, rcond=None)[0]
assert not np.allclose(A @ np.zeros(3), b)
assert not np.allclose(A @ (x3 + x3), b)
print("Ax=b: not a subspace")

# intersection: U1=xy-plane, U2=xz-plane -> x-axis
u = np.array([3, 0, 0])
v = np.array([5, 0, 0])
lam = 4
assert np.array_equal(np.zeros(3)[1:], np.zeros(2))
assert np.array_equal((lam * u)[1:], np.zeros(2))
assert np.array_equal((u + v)[1:], np.zeros(2))
print("intersection: subspace")
```

## Reflection Note

The Ax=0 vs Ax=b contrast was the clearest takeaway. Both use the
same matrix A, but Ax=0 always includes the zero vector while Ax=b
does not. Adding two solutions of Ax=b gives A(x+x) = 2b, which
lands outside the solution set. The failure is immediate once you
write it out.

Figure 2.6 was worth going through carefully. A and C both contain
the origin but still fail closure, which is easy to overlook if the
only check is whether zero is included. The conditions work together,
not independently.

Using assert instead of print made a difference. print just shows
a number to look at, but assert forces each condition to be written
as a testable statement. translating "closed under scalar
multiplication" into actual code and watching it either pass or break
made how it actually works stick in a way that reading the definition
alone does not.
